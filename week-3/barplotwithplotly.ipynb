```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go
```

```python
timesData = pd.read_csv("/kaggle/input/university-rankings-2011/University_Rankings_20220204.csv")
```

```python
timesData.head()
```

```python
df2013 = timesData[timesData.year == 2013].iloc[:3,:] 
df2013
```

```python
trace1 = go.Bar( 
                x = df2013.university_name,
                y = df2013.citations,  
                name = "citations",    
                marker = dict(color = 'rgba(255, 174, 255, 0.5)', 
                             line=dict(color='rgb(0,0,0)',width=1.5)), 
                text = df2013.country) 

trace2 = go.Bar(
                x = df2013.university_name, 
                y = df2013.teaching,   
                name = "teaching",    
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',   
                              line=dict(color='rgb(0,0,0)',width=1.5)), 
                text = df2013.country)  

data = [trace1, trace2] 

layout = go.Layout(barmode = "group") 
                                     

fig = go.Figure(data = data, layout = layout) 

iplot(fig)
```

```python
x = df2013.university_name

trace1 = { 
  'x': x, 
  'y': df2013.citations, 
  'name': 'citation',
  'type': 'bar' 
};
trace2 = {
  'x': x,  
  'y': df2013.teaching,  
  'name': 'teaching',  
  'type': 'bar' 
};

data = [trace1, trace2]; 

layout = {
  'xaxis': {'title': 'Top 3 universities'},
  'barmode': 'relative',  
  'title': 'citations and teaching of top 3 universities in 2013'
};

fig = go.Figure(data = data, layout = layout) 

iplot(fig)
```
